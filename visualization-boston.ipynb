```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# Agregar el nombre de las columnas
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

df = pd.read_csv('housing.csv', header=None, delimiter=r"\s+", names=column_names)
```

```python
df
```

### Variables del dataset

* CRIM: Tasa de criminalidad per cápita por ciudad.
* ZN: Proporción de terreno residencial dividido en zonas para lotes de más de 25,000 pies cuadrados.
* INDUS: Proporción de acres comerciales no minoristas por ciudad.
* CHAS: Variable ficticia Charles River (1 si el tramo limita con el río; 0 en caso contrario).
* NOX: Concentración de óxidos nítricos (partes por 10 millones).
* RM: Número medio de habitaciones por vivienda.
* AGE: Proporción de unidades ocupadas por propietarios construidas antes de 1940.
* DIS: Distancias ponderadas a cinco centros de empleo de Boston.
* RAD: Índice de accesibilidad a carreteras radiales.
* TAX: Tasa de impuesto a la propiedad de valor completo por $10,000.
* PTRATIO: Proporción alumno-maestro por ciudad.

```python
plt.figure(figsize=(10,6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix,annot=True, cmap='viridis')
plt.title('Mapa de calor de correlacion entre variables')
plt.show()
```

```python
plt.hist(df['CRIM'], bins=30)
plt.title('Distribucion por tasa de criminalidad')
plt.xlabel('Tasa de criminalidad per capita')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.boxplot(df['NOX'])
plt.title('Boxplot de oxidos nitricos')
plt.ylabel('Concentracion')
plt.show()
```

```python
plt.hist(df['AGE'], bins=20, edgecolor='black')
plt.title('Distribucion antiguedad de las construcciones')
plt.xlabel('Proporcion de unidades construidas antes de 1940')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.hist(df['RAD'], bins=20, edgecolor='black',color='orange')
plt.title('Accesibilidad a carreteras radiales')
plt.xlabel('Indice de accebilidad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.scatter(df['RAD'],df['MEDV'],alpha=0.5)
plt.title('Carreteras vs. precio medio de las viviendas')
plt.xlabel('Carreteras')
plt.ylabel('Precio medio de las viviendas')
plt.show()
```

```python
plt.scatter(df['RM'],df['MEDV'],alpha=0.5)
plt.title('Numero de habitaciones vs. precio medio de las viviendas')
plt.xlabel('Numero de habitaciones')
plt.ylabel('Precio medio de las viviendas')
plt.show()
```

```python
plt.scatter(df['DIS'],df['MEDV'],alpha=0.5,color='green')
plt.title('Distancia a centros de empleo vs. precio medio de las viviendas')
plt.xlabel('Distancia a centro de empleos')
plt.ylabel('Precio medio de las viviendas')
plt.show()
```

```python
plt.scatter(df['LSTAT'],df['MEDV'],alpha=0.5,color='purple')
plt.title('Porcentaje de "Statusinferior" vs. precio medio de las viviendas')
plt.xlabel('Porcentaje de "Statusinferior"')
plt.ylabel('Precio medio de las viviendas')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.violinplot(x='RAD',y='MEDV',data=df)
plt.title('Valor medio de las viviendas por carretera radial')
plt.xlabel('Indice de acceso a la carretera radial')
plt.ylabel('Valor medio de las viviendas')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.regplot(x='CRIM',y='MEDV',data=df,scatter_kws={"s":20}, line_kws={"color":"red"})
plt.title('Tasa de criminalidad por valor medio de las viviendas')
plt.xlabel('Tasa de criminalidad')
plt.ylabel('Valor medio de las viviendas')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.regplot(x='TAX',y='MEDV',data=df,scatter_kws={"s":20}, line_kws={"color":"green"})
plt.title('Tasa de impuestos a la propiedad por valor medio de las viviendas')
plt.xlabel('Tasa de impuestos')
plt.ylabel('Valor medio de las viviendas')
plt.show()
```

```python
g = sns.PairGrid(df, y_vars=['MEDV'], x_vars=['LSTAT','RM','AGE','DIS','NOX'], height=5)
g.map(sns.regplot,scatter_kws={"s":20},line_kws={"color":"red"})
g.add_legend()
plt.suptitle('Relaciones de variables seleccionadas con el valor medio de las viviendas')
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
selected_features = ['CRIM','RM','AGE','DIS','MEDV']

sns.pairplot(df[selected_features], diag_kind='kde',kind='reg', plot_kws={'line_kws':{'color': 'red'}})
plt.suptitle('Pairplot del dataset Boston')
plt.subplots_adjust(top=0.9)
plt.show()
```

* A mayor criminalidad menor precio de las viviendas
* Numero de habitaciones aumenta el valor de las viviendas
* Las viviendas mas antiguas tienden a tener valores mas bajos
* Accesibilidad: a mayor distancia de los centros de empleo menor valores de las viviendas
